```python
# Merging dataframes in Pandas
import pandas as pd
```

```python
df1 = pd.read_csv(r"/content/LOTR(Python-Pandas dataset 7).csv")
df1
```

```python
df2 = pd.read_csv(r"/content/LOTR 2(Python-Pandas dataset 8).csv")
df2
```

```python
# inner join using merge
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
# outer join using merge
df1.merge(df2, how = 'outer')
```

```python
# left join using merge
df1.merge(df2, how = 'left')
```

```python
# right join using merge
df1.merge(df2, how  = 'right')
```

```python
# cross join using merge
df1.merge(df2, how  = 'cross')
```

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_Left', rsuffix = '_Right')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left', rsuffix = '_Right', how = 'outer')
df4
```

```python
#concat puts first datatable on second datatable
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], join = 'inner', axis = 1)
```

```python
pd.concat([df1,df2], join = 'outer', axis = 1)
```
