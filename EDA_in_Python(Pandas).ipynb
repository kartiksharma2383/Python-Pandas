```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"/content/world_population(Python-Pandas dataset 5).csv")
df
```

```python
# increase the population data by .00 %
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.nunique()
```

```python
#sorting, head(10) will gives top 10 rank values
df.sort_values(by="World Population Percentage", ascending = False).head(10)
```

```python
#correlation(compare every column with every another column and give result of how closely they are)
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True), annot = True)
plt.rcParams['figure.figsize'] = (10,10)
plt.show()
```

```python
#Grouping the data
df.groupby('Continent').mean(numeric_only=True).sort_values(by = "2022 Population", ascending = False)
```

```python
df[df['Continent'].str.contains('Oceania')]
```

```python
df2 = df.groupby('Continent').mean(numeric_only=True).sort_values(by = "2022 Population", ascending = False)
df2
```

```python
df3 = df2.transpose()
df3
```

```python
df3.plot()
```
