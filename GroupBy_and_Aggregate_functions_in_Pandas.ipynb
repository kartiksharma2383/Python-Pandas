```python
#Group By and Aggregate Functions
import pandas as pd
```

```python
df = pd.read_csv("/content/Flavors(Python-Pandas dataset 6).csv")
df
```

```python
group_by_frame = df.groupby('Base Flavor')
```

```python
group_by_frame[['Flavor Rating', 'Texture Rating', 'Total Rating']].mean()
```

```python
df.groupby('Base Flavor').mean(numeric_only = True)
```

```python
df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor').min()
```

```python
df.groupby('Base Flavor').max()
```

```python
df.groupby('Base Flavor').sum()
```

```python
df.groupby('Base Flavor').agg({
'Flavor Rating': ['mean', 'max', 'count', 'sum'],
'Texture Rating': ['mean', 'max', 'count', 'sum']})
```

```python
df.groupby(['Base Flavor', 'Liked'])[['Flavor Rating', 'Texture Rating', 'Total Rating']].mean()
```

```python
df.groupby(['Base Flavor', 'Liked']).agg({'Flavor Rating':{'mean', 'max', 'count', 'sum'}})
```

```python
df.groupby('Base Flavor').describe()
```
