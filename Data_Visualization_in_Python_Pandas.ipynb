```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Ice Cream Ratings(Python-Pandas dataset 9).csv')
df = df.set_index('Date')
df
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Rating', ylabel = 'Scores')
```

```python
df.plot(kind = 'line', subplots = True)
```

```python
df.plot(kind = 'bar', stacked = True)
```

```python
df['Flavor Rating'].plot(kind = 'bar', stacked = True)
```

```python
df.plot.barh(stacked = True)
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', c = 'red')
```

```python
df.plot.hist(bins = 10)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10,5))
```

```python
df.plot.pie(y = 'Flavor Rating', figsize = (5,5))
```
