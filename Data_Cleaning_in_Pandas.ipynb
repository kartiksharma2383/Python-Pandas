```python
import pandas as pd
```

```python
df = pd.read_excel(r"/content/Customer Call List(Python-Pandas dataset 10).xlsx")
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
#df["Last_Name"] = df["Last_Name"].str.lstrip("...")
#df["Last_Name"] = df["Last_Name"].str.lstrip("/")
#df["Last_Name"] = df["Last_Name"].str.rstrip("_")
#or
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

```python
df["Phone_Number"] = df["Phone_Number"].astype(str).str.replace(r'[Nn]/a|NaN|-|/|\||nan|Na', '', regex=True)
df["Phone_Number"] = df["Phone_Number"].str.replace(r'(\d{3})(\d{3})(\d{4})', r'\1-\2-\3', regex=True)
df
```

```python
df[["Stree Address", "State", "Zip_Code"]] = df["Address"].str.split(', ', n=2, expand = True)
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes', 'Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No', 'N')
df
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes', 'Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No', 'N')
df
```

```python
df = df.fillna('')
df
```

```python
for x in df.index:
  if df.loc[x, "Do_Not_Contact"] == 'Y':
    df.drop(x, inplace = True)
df
```

```python
for x in df.index:
  if df.loc[x, "Phone_Number"] == '':
    df.drop(x, inplace = True)
df
```

```python
df = df.reset_index(drop = True)
df
```
