```python
import pandas as pd
```

```python
df = pd.read_csv(r'/content/world_population(Python-Pandas dataset 5).csv')
df
```

```python
#Filtering the data
df[df['Rank'] < 10]
```

```python
specific_countries = ['Bangladesh', 'Brazil']
df[df['Country'].isin(specific_countries)]
```

```python
df[df['Country'].str.contains('United')]
```

```python
df2 = df.set_index('Country')
df2
```

```python
df2.filter(items = ['Continent', 'CCA3'], axis = 1)
```

```python
df2.filter(items = ['Zimbabwe'], axis = 0)
```

```python
df2.filter(like = 'United', axis = 0)
```

```python
df2.loc['United States']
```

```python
df2.iloc[3]
```

```python
#Order By
df[df['Rank'] < 10].sort_values(by=['Rank', 'Country'], ascending = True)
```

```python
df[df['Rank'] < 10].sort_values(by=['Rank', 'Country'], ascending = False)
```

```python
df[df['Rank'] < 10].sort_values(by=['Continent', 'Country'], ascending = [False, True])
```

```python
#Indexing
df = pd.read_csv(r'/content/world_population(Python-Pandas dataset 5).csv', index_col = "Country")
df
```

```python
df.reset_index(inplace = True)
df
```

```python
df.set_index('Country', inplace = True)
df
```

```python
df.loc['Albania']
```

```python
df.iloc[1]
```

```python
df.reset_index(inplace = True)
df.set_index(['Continent','Country'], inplace = True)
df
```

```python
df.sort_index(ascending = True)
```

```python
df.sort_index(ascending = False)
```
